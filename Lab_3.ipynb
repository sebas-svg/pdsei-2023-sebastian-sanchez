```python
import cv2
import numpy as np
import urllib.request
from matplotlib import pyplot as plt

# URL de la imagen en GitHub
url = "https://raw.githubusercontent.com/sebas-svg/pdsei-2023-sebastian-sanchez/main/Tarea3/solvay.jpg"

# Descarga la imagen desde la URL
response = urllib.request.urlopen(url)
img_array = np.asarray(bytearray(response.read()), dtype=np.uint8)

# Guarda la imagen temporalmente
with open('temp_image.jpg', 'wb') as f:
    f.write(img_array)

# Lee la imagen con OpenCV
image = cv2.imread('temp_image.jpg', 0)

# Comprueba si la imagen se leyó correctamente
if image is not None:
    # Aplica la función T a la imagen
    A = 75
    B = 125
    H = 225
    T = np.zeros((256), dtype=np.uint8)

    for i in range(256):
        if i <= A:
            T[i] = i
        elif i > A and i < B:
            T[i] = H
        elif i >= B:
            T[i] = i

    # Aplica la transformación a la imagen
    transformed_image = cv2.LUT(image, T)

    # Muestra la imagen original y la imagen transformada
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), cmap='gray')
    plt.axis('off')
    plt.title('Imagen Original')

    plt.subplot(1, 2, 2)
    plt.imshow(transformed_image, cmap='gray')
    plt.axis('off')
    plt.title('Imagen con Transformación T')

    plt.show()
else:
    print("No se pudo cargar la imagen.")
```

```python
# Importamos librerías:

import cv2
import matplotlib.pyplot as plt

import numpy as np

# Imágenes en escala de gris:

# URL de las imagenes en GitHub
url1 = "https://raw.githubusercontent.com/sebas-svg/pdsei-2023-sebastian-sanchez/main/Tarea3/chisco.jpg"
url2 = "https://raw.githubusercontent.com/sebas-svg/pdsei-2023-sebastian-sanchez/main/Tarea3/Rupicola.jpg"

# Descarga las imagenes desde la URL
response1 = urllib.request.urlopen(url1)
response2 = urllib.request.urlopen(url2)

img_array1 = np.asarray(bytearray(response1.read()), dtype=np.uint8)
img_array2 = np.asarray(bytearray(response2.read()), dtype=np.uint8)

# Guarda las imagenes temporalmente
with open('temp_image1.jpg', 'wb') as f:
    f.write(img_array1)
with open('temp_image2.jpg', 'wb') as f:
    f.write(img_array2)

# Lee la imagen con OpenCV
imagen_1_gray = cv2.imread('temp_image1.jpg',0)
imagen_2_gray = cv2.imread('temp_image2.jpg',0)

# Comprueba si las imagenes se leyeron correctamente
if imagen_1_gray is not None:
    # Muestra la imagen utilizando matplotlib
    plt.imshow(cv2.cvtColor(imagen_1_gray, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Desactiva los ejes
    plt.show()
else:
    print("No se pudo cargar la imagen 1.")

if imagen_2_gray is not None:
    # Muestra la imagen utilizando matplotlib
    plt.imshow(cv2.cvtColor(imagen_2_gray, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Desactiva los ejes
    plt.show()
else:
    print("No se pudo cargar la imagen 2.")
```

```python
#Obtengo las equalizaciones de las imágenes:

equ_imagen_1_gray = cv2.equalizeHist(imagen_1_gray)

equ_imagen_2_gray = cv2.equalizeHist(imagen_2_gray)

# Imprimimos las imágenes, sus histogramas y los histogramas acumulados de la imágenes y sus equalizaciones:

plt.figure(figsize=(15,15))

plt.subplot(421); plt.imshow(imagen_1_gray, cmap="gray"); plt.title("Imagen 1 GRIS")
plt.subplot(422); plt.imshow(imagen_2_gray, cmap="gray"); plt.title("Imagen 2 GRIS")
plt.subplot(423); plt.hist(imagen_1_gray.ravel(),256, [0,256]); plt.title("Histograma de la Imagen 1")

plt.subplot(424); plt.hist(imagen_2_gray.ravel(),256, [0,256]); plt.title("Histograma de la Imagen 2")

plt.subplot(425); plt.hist(imagen_1_gray.ravel(),256, [0,256], cumulative=True);

plt.title("Histograma acumulado de la Imagen 1")
```

```python
def histogram_matching(source_image, reference_image):
    source_hist = cv2.calcHist([source_image], [0], None, [256], [0, 256])
    reference_hist = cv2.calcHist([reference_image], [0], None, [256], [0, 256])

    source_hist_cumsum = np.cumsum(source_hist) / source_image.size
    reference_hist_cumsum = np.cumsum(reference_hist) / reference_image.size

    mapping = np.interp(source_hist_cumsum, reference_hist_cumsum, range(256))
    matched_image = cv2.LUT(source_image, mapping.astype('uint8'))

    return matched_image
```

```python
source_image=imagen_1_gray
reference_image=imagen_2_gray

# Aplicar la especificación de histograma
matched_image = histogram_matching(source_image, reference_image)

# Mostrar y guardar las imágenes de entrada y salida

plt.imshow(cv2.cvtColor(matched_image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Desactiva los ejes
plt.show()

# Guardar la imagen de salida
cv2.imwrite('matched_image.jpg', matched_image)
```
